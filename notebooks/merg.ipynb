```python
import pandas as pd

def load_and_merge_data(gen_file, weather_file):
    gen_df = pd.read_csv(gen_file)
    weather_df = pd.read_csv(weather_file)

    # Convert to datetime
    gen_df['DATE_TIME'] = pd.to_datetime(gen_df['DATE_TIME'])
    weather_df['DATE_TIME'] = pd.to_datetime(weather_df['DATE_TIME'])

    # Merge on DATE_TIME
    merged_df = pd.merge(gen_df, weather_df, on='DATE_TIME')

    # Optional: drop duplicates and nulls
    merged_df = merged_df.drop_duplicates().dropna()

    # Save merged
    merged_df.to_csv('data/processed/merged_cleaned_data.csv', index=False)
    print("✅ Merged and saved:", merged_df.shape)

    return merged_df

# Example usage:
load_and_merge_data("data/raw/Plant_1_Generation_Data.csv", "data/raw/Plant_1_Weather_Sensor_Data.csv")
```
